```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = 'https://your.gg/ko/kr/profile/T1%20Gumayusi'
resp = requests.get(url)

soup = BeautifulSoup(resp.text)
```

```python
 

title_list=[]
point_list=[]



# <span size="11" class="buiiz7-0 sc-1vxty03-3 hZCTWr swzFw">인분</span>
# <div class="sc-1vxty03-2 ipiPoD">
# <span size="11" class="buiiz7-0 sc-1vxty03-3 hZCTWr swzFw">인분</span>
# <span size="16" class="buiiz7-0 sc-1vxty03-4 ikoBPa">0.76</span></div>
for line in soup.findAll('div', class_="sc-1vxty03-2 ipiPoD"):
    title = line.find('span', class_="buiiz7-0 sc-1vxty03-3 hZCTWr swzFw") 
    if title:
#         print('... ->', title.get_text())
        title_list.append(title.get_text())

    point = line.find('span', class_="buiiz7-0 sc-1vxty03-4 ikoBPa")
    if point:
#         print('point --> ', point.get_text())
        point_list.append(point.get_text())
    
    
            
```
