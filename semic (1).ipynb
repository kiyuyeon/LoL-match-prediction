```python
    import requests
    from bs4 import BeautifulSoup
    import pandas as pd
    from selenium import webdriver
    from webdriver_manager.chrome import ChromeDriverManager
    from selenium.webdriver.chrome.service import Service
    from selenium.webdriver.common.by import By
    from selenium.webdriver.chrome.options import Options

    import time
```

```python
url = 'https://your.gg/ko/kr/profile/T1%20Gumayusi'
resp = requests.get(url)

soup = BeautifulSoup(resp.text)
```

```python

point_list=[]

# //*[@id="__next"]/div/main/div[2]/div/div/div[1]/div[1]/button[2]

path = "C:/chromedriver_win32/chromedriver.exe"
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))
driver.get(url)

xpath = '//*[@id="__next"]/div/main/div[2]/div/div/div[1]/div[1]/button[2]'
sole=driver.find_element(By.XPATH, xpath)
sole.click()
time.sleep(1)
soup = BeautifulSoup(driver.page_source)

ypath = '//*[@id="__next"]/div/main/div[2]/div/div/div[1]/div[2]/button[5]'
bottom=driver.find_element(By.XPATH, ypath)
bottom.click()
time.sleep(1)
soup = BeautifulSoup(driver.page_source)

for line in soup.findAll('div', class_="sc-1vxty03-2 ipiPoD"):
    point = line.find('span', class_="buiiz7-0 sc-1vxty03-4 ikoBPa")
    if point:
        # print('point --> ', point.get_text())
        point_list.append(point.get_text())
        
        
sr1 = point_list[0::3]
sr2 = point_list[1::3]
sr3 = point_list[2::3]

df = pd.DataFrame()   
df['인분'] = point_list[0::3]#sr1
df['라인전'] = point_list[1::3]#sr2
df['팀운'] = point_list[2::3]#sr3
            
df



```

```python
column_name = ['인분', '라인전','팀운']
df= pd.DataFrame(columns=column_name)
index = 0

#'div' 전체 태그 검색
for line in soup.find_all('div',class_='sc-1vxty03-0 fWWOpD d490eh-7 bKpSsu'):
    # 인분, 라인전, 팀운 태그만 검색
    for line_1 in line.find_all('div',class_='sc-1vxty03-1 frbyeq'):
        line_span = line_1.find_all('span')
        column_value = []
        for i in range(len(line_span)):
            if line_span[i]:
                text = line_span[i].get_text().strip('\n')
                if text not in column_name:
                    column_value.append(text)
        df.loc[index] = column_value
        index += 1

df
```

```python

# cs 킬뎃 킬관여율%
point_list=[]


for line in soup.findAll('div', class_="sc-1p49g2y-0 kknGKj d490eh-9 iTqkMo"):
    for item in line.findAll('span'):
        #print('point --> ', item.get_text())
        point_list.append(item.get_text())
print(point_list)
sr1 = point_list[0::4]
sr2 = point_list[2::4]
sr3 = point_list[3::4]

df2 = pd.DataFrame()   
df2['point'] = sr1
df2['cs'] = sr2
df2['kp'] = sr3
            
```

```python
df_test = pd.concat([df, df2], axis=1)
df_test.to_csv('원딜데이터.csv',index=False,encoding = 'cp949')#중간저장
```
